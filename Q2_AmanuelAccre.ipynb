```python
accre_select_df=accre_df2.loc[:,['JobID','User','Submit','Start','job_duration']]
```

# sort the the accre_select_df by job_duration and user name in ascending order.

```python
accre_select_df=accre_select_df.sort_values(by=['job_duration', 'Submit','Start','User',],ascending=True)
```

```python
accre_select_df.head()
```

```python
month=pd.DataFrame(accre_select_df['Submit'].dt.month)
month.rename(columns={'Submit':'month'},inplace=True)
```

```python
date=pd.DataFrame(accre_select_df['Submit'].dt.date)
date.rename(columns={'Submit':'date'},inplace=True)
```

```python
hour=pd.DataFrame(accre_select_df['Submit'].dt.hour)
hour.rename(columns={'Submit':'hour'},inplace=True)
```

```python
accre_select_df.head()
```

```python
accre_select_df1=accre_select_df.join(month)
```

```python
accre_select_df1.head()
```

```python
accre_select_df2=accre_select_df1.join(date)
```

```python
accre_select_df2=accre_select_df2.join(hour)
```

```python
accre_select_df2.head()
```

```python
accre_select_df2.head()
```

```python
accre_select_df2['hour'].dtype
```

```python
accre_5_minutes= accre_select_df2[accre_select_df2.job_duration<'0 days 00:05:00']
```

```python
accre_5_minutes.head()
```

```python
accre_5_minutes=accre_5_minutes.sort_values(by=['Submit','month','date','hour','job_duration','User',],ascending=True)
```

```python
count_5_minutes_job=accre_5_minutes.groupby(['User','month','date','hour'])['JobID'].count()
```

```python
count_5_minutes_job.head()
```

# users sumbitted more than 500 jobs each taking shorter than 5 minutes grouped by month,date and hour

```python

count_5_minutes_job[count_5_minutes_job> 500]
```
